```python
import pandas as pd
import pyreadstat

# 1. Escribir el nombre del archivo .sav
nombre_archivo_sav = 'Casen 2015.sav'

# 2. Eligir el nombre para el nuevo archivo .csv
nombre_archivo_csv = 'datos_convertidos2015.csv'

# 3. Leer el archivo .sav y guardarlo como DataFrame
df, meta = pyreadstat.read_sav(nombre_archivo_sav)

# 4. Guardar el DataFrame como un archivo .csv
df.to_csv(nombre_archivo_csv, index=False)

print(f"¡Éxito! Archivo convertido y guardado como: {nombre_archivo_csv}")
```

```python
nombre_archivo = 'datos_convertidos2015.csv'

# Escribir los nombres exactos de las columnas que necesitamos
columnas_que_me_interesan = [
    "id_vivienda",
    'region',
    "provincia",
    'comuna',
    'zona',
    'ypchtot',
    'pobreza'
]

try:
    # 3. Cargar el CSV usando 'usecols' para filtrar
    df = pd.read_csv(
        nombre_archivo,
        usecols=columnas_que_me_interesan
    )
    
    # 4. Mostrar un resumen para verificar que funcionó
    print(f"¡Éxito! Se cargó el DataFrame solo con {len(df.columns)} columnas.")
    print("\n--- Primeras 5 filas: ---")
    print(df.head())
    
    print("\n--- Información del DataFrame: ---")
    df.info()

except FileNotFoundError:
    print(f"Error: No se encontró el archivo '{nombre_archivo}'.")
    print("Asegúrate de que esté en la misma carpeta que tu notebook.")
except ValueError as e:
    print(f"Error: {e}")
    print("\nPosible causa: Algún nombre en 'columnas_que_me_interesan' no existe en el archivo CSV.")
    print("Revisa que los nombres de las columnas estén escritos exactamente igual.")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```
