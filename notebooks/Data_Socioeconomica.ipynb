## Datos combinados

```python
# --- CELDA ACTUALIZADA: CONSOLIDACIÓN DE DATOS SOCIOECONÓMICOS ---

# 1. Definir los archivos .sav y años correspondientes
archivos_casen_sav = {
    2015: 'Casen 2015.sav',
    2017: 'Casen 2017.sav',
    2022: 'Base de datos Casen 2022 SPSS_18 marzo 2024.sav'
}
# Archivo adicional para las comunas de 2022
archivo_comunas_2022 = 'Base de datos provincia y comuna Casen 2022 SPSS.sav'


lista_df = []
ruta_base = os.path.join('..', 'data', 'raw')

# 2. Cargar cada archivo .sav, seleccionando y estandarizando las columnas necesarias
print("Cargando y procesando archivos CASEN .sav...")
for anio, nombre_archivo in archivos_casen_sav.items():
    try:
        ruta_completa = os.path.join(ruta_base, nombre_archivo)
        df_temp, meta = pyreadstat.read_sav(ruta_completa, apply_value_formats=True, formats_as_category=False)
        
        df_temp['Anyo'] = anio
        
        # --- LÓGICA ESPECIAL PARA 2022 ---
        if anio == 2022:
            print(" - Procesando archivo adicional de comunas para 2022...")
            ruta_comunas = os.path.join(ruta_base, archivo_comunas_2022)
            df_comunas, meta_comunas = pyreadstat.read_sav(ruta_comunas, usecols=['folio', 'comuna'])
            
            # Unir el dataframe principal con el de comunas usando 'folio'
            df_temp = pd.merge(df_temp, df_comunas, on='folio', how='left')
            print("   - Fusión de comunas completada.")
        
        # Renombrar 'area' a 'zona' en el dataset de 2022 para consistencia
        if 'area' in df_temp.columns:
            df_temp.rename(columns={'area': 'zona'}, inplace=True)
            
        # Seleccionar solo las columnas necesarias. Mantenemos 'comuna' porque es
        # CRUCIAL para la fusión con los datos de energía.
        columnas_necesarias = ['region', 'comuna', 'zona', 'ytot', 'pobreza', 'Anyo']
        df_temp = df_temp[columnas_necesarias]
        
        lista_df.append(df_temp)
        print(f" - Archivo '{nombre_archivo}' procesado.")
    except FileNotFoundError:
        print(f" - ADVERTENCIA: No se encontró el archivo '{nombre_archivo}'. Será omitido.")
    except Exception as e:
        print(f" - ERROR procesando '{nombre_archivo}': {e}")

if not lista_df:
    print("\nError: No se pudo cargar ningún archivo CASEN. El proceso se ha detenido.")
else:
    df_casen_full = pd.concat(lista_df, ignore_index=True)
    print("\nArchivos combinados exitosamente.")

    # 3. Filtrar por Región Metropolitana
    print("Filtrando por Región Metropolitana...")
    df_rm = df_casen_full[df_casen_full['region'].str.contains("Metropolitana", na=False)].copy()

    # 4. Calcular indicadores agregados por comuna y año
    print("Calculando indicadores socioeconómicos por comuna y año...")
    
    # Crear indicadores binarios para pobreza y zona urbana
    df_rm['es_pobre'] = df_rm['pobreza'].str.contains("Pobre", na=False).astype(int)
    df_rm['es_urbano'] = (df_rm['zona'] == 'Urbano').astype(int)

    # Agrupar por año y nombre de comuna para calcular las métricas
    df_agregado = df_rm.groupby(['Anyo', 'comuna']).agg(
        # Usar 'ytot' para el ingreso total corregido del hogar
        Ingreso_Promedio_Hogar=('ytot', 'mean'),
        Indice_Pobreza_pct=('es_pobre', lambda x: x.mean() * 100 if len(x) > 0 else 0),
        Pct_Poblacion_Urbana=('es_urbano', lambda x: x.mean() * 100 if len(x) > 0 else 0)
    ).reset_index()

    # 5. Renombrar columnas para que coincidan con el script de limpieza
    print("Renombrando columnas para compatibilidad...")
    df_final_socio = df_agregado.rename(columns={
        'comuna': 'Nombre Comuna',
        'Indice_Pobreza_pct': 'Indice Pobreza Multidimensional (%)'
    })
    
    # 6. Guardar el archivo CSV unificado en el formato requerido
    ruta_destino = os.path.join('..', 'data', 'raw', 'datos_socioeconomicos.csv')
    try:
        df_final_socio.to_csv(ruta_destino, index=False, sep=';', encoding='latin1')
        print(f"\n¡Éxito! Archivo consolidado guardado en: {ruta_destino}")
        print("\n--- Vista previa de los datos generados: ---")
        display(df_final_socio.head())
    except Exception as e:
        print(f"\nError al guardar el archivo final: {e}")
```

```python
import pandas as pd
import os
import numpy as np

# --- Rutas de Archivos ---

# Tu ruta local al archivo XLSX (ajusta la ruta si es necesario)
# Asegúrate de que este path apunte al archivo .xlsx real.
path3 = os.path.join('..', 'data','raw', 'Libro de codigos Base de datos provincia y comuna Casen 2022.xlsx')

# La ruta a tu archivo de datos socioeconómicos
df_socioeconomicos_path = "datos_socioeconomicos.csv"


# --- 1. Cargar y Limpiar el Libro de Códigos (XLSX) ---

# Leemos el Excel sin encabezados (header=None) para poder seleccionar las columnas
# de código y nombre de comuna por índice.
df_codigos = pd.read_excel(path3, header=None)

# Basado en la estructura del archivo, la información de códigos está en la columna 3 (código)
# y columna 4 (nombre), y los datos reales empiezan en la fila 7 (índice 6).
df_codigos_clean = df_codigos.iloc[6:, [3, 4]].copy()
df_codigos_clean.columns = ['codigo_comuna', 'nombre_comuna']

# Limpieza de la columna de códigos
df_codigos_clean.dropna(subset=['codigo_comuna'], inplace=True)
df_codigos_clean['codigo_comuna'] = pd.to_numeric(
    df_codigos_clean['codigo_comuna'], errors='coerce'
).astype('Int64')
df_codigos_clean.dropna(subset=['codigo_comuna'], inplace=True)

# Crear el diccionario de mapeo: código -> nombre
mapa_comunas = df_codigos_clean.set_index('codigo_comuna')['nombre_comuna'].to_dict()

condicion_2022 = df_final_socio['Anyo'] == 2022

# Convertir la columna 'Nombre Comuna' a entero para el mapeo (solo para las filas de 2022)
df_final_socio.loc[condicion_2022, 'Codigo_Comuna_Int'] = pd.to_numeric(
    df_final_socio.loc[condicion_2022, 'Nombre Comuna'], errors='coerce'
).astype('Int64')

# Mapear los códigos a los nombres en una columna temporal
df_final_socio.loc[condicion_2022, 'Nombre Comuna_Actualizado'] = (
    df_final_socio.loc[condicion_2022, 'Codigo_Comuna_Int']
    .map(mapa_comunas)
)

# Reemplazar los códigos en la columna original 'Nombre Comuna' con los nombres actualizados
df_final_socio.loc[
    condicion_2022 & df_final_socio['Nombre Comuna_Actualizado'].notna(),
    'Nombre Comuna'
] = df_final_socio['Nombre Comuna_Actualizado']

# Eliminar las columnas temporales
df_final_socio.drop(columns=['Codigo_Comuna_Int', 'Nombre Comuna_Actualizado'], inplace=True)

ruta_destino = os.path.join('..', 'data', 'raw', 'datos_socioeconomicos2.csv')
# --- 3. Guardar el Resultado ---
try:
    df_final_socio.to_csv(ruta_destino, index=False, sep=';', encoding='latin1')
    print(f"\n¡Éxito! Archivo consolidado guardado en: {ruta_destino}")
finally:
    df_final_socio.sample(20)
```
